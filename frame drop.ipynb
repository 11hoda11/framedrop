### Finding the video duration with "videofileclip"

```python
from moviepy.editor import VideoFileClip

def calculate_duration(video_path):
    # Load the video clip
    clip = VideoFileClip(video_path)

    # Get the duration in seconds
    duration = clip.duration

    # Release the video clip
    clip.close()

    # Return the duration
    return duration

# Specify the path to the video file
#video_path = "C:/Users/hodan/OneDrive/Desktop/Video codes/VID-20200320-WA0014_2.mp4"
video_path = "Z:/CAMERA/CAMERA Visits/Mobility Visit/Study Subjects/20191112/9769/Video_Data/CH_002/sess_1_gait_back_forth_105417_105435.mp4"

# Call the function to calculate the duration
video_duration = calculate_duration(video_path)

# Print the result
print("Video duration:", video_duration, "seconds")
```

### Define a function to extraxt timestamp of each frame



```python
import cv2
import numpy as np

def extract_frame_timestamps(video_path):
    # Open the video file
    video = cv2.VideoCapture(video_path)

    # Check if the video file was successfully opened
    if not video.isOpened():
        print("Could not open the video file.")
        return

    timestamps = []
    others = []
    while True:
        # Read the frame and timestamp
        ret, frame = video.read()

        # Check if frame was read successfully
        if not ret:
            break
         
        # Extract the timestamp from the frame
        timestamp = video.get(cv2.CAP_PROP_POS_MSEC) / 1000.0  # Convert to seconds
        timestamps.append(timestamp)
        
    # Release the video file
    video.release()

    return timestamps 

```

### Plot timestamps for all views

```python
import matplotlib.pyplot as plt

# Generate a list of unique video IDs in the DataFrame
video_ids = df['Video'].unique()

# Define a color map
colors = plt.cm.rainbow  # You can choose any colormap

# Plot timestamp vs. frame for each video with different colors
for video_id in video_ids:
    video_data = df[df['Video'] == video_id]
    frame_indices = video_data['Frame']
    timestamps = video_data['Timestamp']
    plt.plot(frame_indices, timestamps, color=colors(video_id / len(video_ids)), label=f"Video {video_id}")

# Add labels and legend
plt.xlabel('Frame')
plt.ylabel('Timestamp')
plt.title('Timestamp vs. Frame for Each Video')
plt.legend()

# Show the plot
plt.show()

```
